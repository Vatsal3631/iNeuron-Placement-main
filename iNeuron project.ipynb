# **Support Vector Machine Classifier**

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
clf = make_pipeline(StandardScaler(), SVC(C=1.0, gamma='auto'))
clf.fit(x_train, y_train)
pickle.dump(clf,open('svm_model.pkl','wb'))
print("Train set mean Accuracy:",clf.score(x_train,y_train))
print("Test set mean Accuracy:",clf.score(x_test,y_test))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
y_pred_clf = clf.predict(x_test)
print("Accuracy: \n", accuracy_score(y_test, y_pred_clf))
print("Confusion Matrix: \n", confusion_matrix(y_test, y_pred_clf))
print("Classification Report: \n", classification_report(y_test, y_pred_clf))
```
